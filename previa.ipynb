## Qual é a região de cada código?

Inferir pelo nome de escola e sua região

```python
def escolasComNome(nome, colunas=['NO_ENTIDADE', 'CO_REGIAO', 'CO_MICRORREGIAO']):
    return escolas[escolas.NO_ENTIDADE.str.contains(nome)][colunas]

escolasComNome("CAXIAS DO SUL|CEFET")
```

```python
escolasComNome("DOM CABRAL")
```
